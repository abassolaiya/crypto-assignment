```python
import numpy as np
import pandas as pd
import requests
import json
from pandas import json_normalize

url = "https://min-api.cryptocompare.com/data/v2/histoday?fsym=BTC&tsym=USD&limit=30"
JSONContent = requests.get(url).json()
content = json.dumps(JSONContent, indent = 4, sort_keys=True)
print(content)
```

```python
from pandas import json_normalize
info = json.loads(content)

df = json_normalize(info['Data']['Data'])
```

```python
df.describe()
```

```python
import time
t = []
for i in df['time']:
    #ctime(df['time'])
    t.append(time.gmtime(i))
df['time1'] = t
```

```python
#https://min-api.cryptocompare.com/data/pricemulti?fsyms=BTC,ETH&tsyms=USD,EUR
url = "https://min-api.cryptocompare.com/data/v2/histoday?fsym=ETH&tsym=USD&limit=30"
JSONContent = requests.get(url).json()
content1 = json.dumps(JSONContent, indent = 4, sort_keys=True)
print(content1)
```

```python
from pandas import json_normalize
info = json.loads(content1)

df1 = json_normalize(info['Data']['Data'])
```

```python
import time
t = []
for i in df1['time']:
    #ctime(df['time'])
    t.append(time.gmtime(i))
df1['time1'] = t
```

```python
df1.head()
```

```python
df['daily returns'] = (df['close']-df['open'])*100/df['open']
df['max returns'] = (df['high']-df['low'])*100/df['open']
df.head()
```

```python
df1['daily returns'] = (df1['close']-df1['open'])*100/df1['open']
df1['max returns'] = (df1['high']-df1['low'])*100/df1['open']
df1.head()
```

```python
BTC_total = df['daily returns'].sum()
BTC_max = df['max returns'].sum()
eth_total = df1['daily returns'].sum()
eth_max = df1['max returns'].sum()
```

```python
print('BTC_total - ' + str(BTC_total))
print('BTC_max - '+ str(BTC_max))
print('eth_total - ' + str(eth_total))
print('eth_max - ' + str(eth_max))
```

#### BTC performd better on the average than ETH in the past 30 days condisdering the daily returns and maximum profit possible

```python
BTC_total_mean = df['daily returns'].mean()
BTC_max_mean = df['max returns'].mean()
eth_total_mean = df1['daily returns'].mean()
eth_max_mean = df1['max returns'].mean()
```

```python
print('BTC_total_mean -' + str(BTC_total_mean))
print('BTC_max_mean -' + str(BTC_max_mean))
print('eth_total_mean -' + str(eth_total_mean))
print('eth_max_mean -' + str(eth_max_mean))
```

```python
print('BTC_total_median -' + str(BTC_total_median))
print('BTC_max_median -' + str(BTC_max_median))
print('eth_total_median -' + str(eth_total_median))
print('eth_max_median -' + str(eth_max_median))
```
